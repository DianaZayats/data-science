Імпорт бібліотек

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

Створення набору даних у вигляді тензорів

```python
# Створення тензорів
scalar = tf.constant(5)  # Скаляр
vector = tf.constant([1, 2, 3])  # Вектор
matrix = tf.constant([[1, 2], [3, 4]])  # Матриця
tensor = tf.constant(np.random.rand(3, 3, 3))  # Тензор 3x3x3

# Виведення створених тензорів
print("Скаляр:", scalar)
print("Вектор:", vector)
print("Матриця:\n", matrix)
print("Тензор:\n", tensor)
```

CRUD-операції над тензорами

```python
# Create: створення нового тензора
new_tensor = tf.constant(np.random.randint(0, 10, (4, 4)))

# Read: читання елементів
print("Елемент з позиції [0, 0]:", new_tensor[0, 0])
print("Перший рядок:", new_tensor[0])

# Update: оновлення значення (неможливо напряму змінити Tensor, тому створимо новий)
updated_tensor = tf.Variable(new_tensor)
updated_tensor[0, 0].assign(99)
print("Оновлений тензор:\n", updated_tensor)

# Delete: видалення рядка/стовпця (імітація через виключення елемента)
reduced_tensor = tf.concat([new_tensor[1:], tf.zeros([1, 4], dtype=tf.int32)], axis=0)
print("Тензор після видалення рядка:\n", reduced_tensor)
```

Класифікація

```python
from sklearn.datasets import make_classification

# Генерація даних для класифікації
X_class, y_class = make_classification(
    n_samples=1000,           # Кількість записів
    n_features=5,             # Загальна кількість ознак
    n_informative=3,          # Кількість інформативних ознак
    n_redundant=1,            # Кількість надлишкових ознак
    n_classes=3,              # Кількість класів
    n_clusters_per_class=1,   # Кількість кластерів на кожен клас
    random_state=42
)

# Виведення перших 5 записів
print("Класифікаційні дані (перші 5 рядків):")
print(X_class[:5])
print("Мітки класів (перші 5):", y_class[:5])

# Візуалізація класифікаційних даних
plt.figure(figsize=(8, 6))
plt.scatter(X_class[:, 0], X_class[:, 1], c=y_class, cmap='viridis', edgecolor='k')
plt.title('Класифікаційні дані')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Class Label')
plt.show()
```

Лінійна регресія

```python
# Генерація даних для лінійної регресії
np.random.seed(42)

# Параметри лінійної моделі: y = m * x + b + noise
n_samples = 100
x = np.random.rand(n_samples) * 10  # Значення ознаки x
m = 2.5                             # Коефіцієнт нахилу
b = 5                               # Зміщення
noise = np.random.normal(0, 1, n_samples)  # Шум

# Формула для генерації значень цільової змінної y
y = m * x + b + noise

# Візуалізація даних
plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.7, edgecolor='k')
plt.title('Лінійна регресія: згенеровані дані')
plt.xlabel('Feature (x)')
plt.ylabel('Target (y)')
plt.grid()
plt.show()
```
